```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('new_ds.csv',index_col=None)
```

```python
# количество уникальных значений в каждом столбце
df_uniq = df.nunique()
df_uniq = df_uniq.sort_values(ascending=True)
plt.figure(figsize=(10,5))

df_uniq.plot(kind='bar')  
```

```python
df_uniq_df = df_uniq.reset_index()
df_uniq_df.columns = ['Column Name', 'Unique Values']

df_uniq_df.to_csv('unique_values.csv', index=False)
```

```python
df_isna = df.isna().sum()
df_isna = df_isna.sort_values(ascending=True)
plt.figure(figsize=(10,5))

df_isna.plot(kind='bar')  
df_isna_df = df_isna.reset_index()
df_isna_df.columns = ['Column Name', 'Isna']

df_isna_df.to_csv('isna_values.csv', index=False)
```

```python
# сгруппируем по инн
result = df.groupby('client_inn').agg(list).reset_index()
```

```python
result.to_csv('inn.csv',index=False)
```
